most popular songs

```python
#quary
most_popular=dft.query("popularity>90", inplace=False).sort_values("popularity", ascending=False)
most_popular[:4]
```

```python
#changing the index from 1,2,3 to release_date

dft.set_index("release_date", inplace=True)
dft.index=pd.to_datetime(dft.index)
dft.head()
```

```python
#artist name at particular index

dft[["artists"]].iloc[25]
```

```python
# changing duration from MS to Sec

dft["duration"]=dft["duration_ms"].apply(lambda x: round(x/1000))

dft.drop("duration_ms", inplace=True, axis=1)
```

```python
dft.duration.head(5)
```

```python
dft.head()
```

Correlation

```python
corr_dft=dft.drop(["key","mode","explicit"],axis=1).corr(method="pearson") #droping few col and use pearson corelation method
```

```python
corr_dft.head(2)
```

```python
plt.figure(figsize=(16,9))
heatmap=sns.heatmap(corr_dft,cmap="inferno",annot=True,linewidths=1, linecolor="white")
heatmap.set_title("corralation heatmap")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
#creating a sample dataset from our dataset of 0.4% of the size
sample_dft=dft.sample(int(0.004*len(dft)))
```

```python
sample_dft.shape
```

```python
dft.shape
```

regression plot betwwen loudness and energy

```python
#as in the heatmap we get positive correlation between enery and loudness

plt.figure(figsize=(10,6))
sns.regplot(data=sample_dft,y="loudness",x="energy",color="blue",line_kws={"color":"black"}).set(title="loudness vs energy")
```

popularity vs acousticness

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_dft,y="popularity",x="acousticness",color="c",line_kws={"color":"black"}).set(title="popularity vs acousticness")
```

total number of songs since 1922

```python
# creating new col name year

dft["dates"]=dft.index.get_level_values("release_date")
dft.dates=pd.to_datetime(dft.dates)
years=dft.dates.dt.year
```

```python
#total number of songs since 1922
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist")
```

Duration of songs

```python
total_dr=dft.duration
fig_dims=(18,7)
fig, ax= plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years, y=total_dr, ax=ax, errwidth=False).set(title="year vs duration")
plt.xticks(rotation=90)
```

```python
sns.set_style(style="whitegrid")
fig, ax =plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years,y=total_dr,ax=ax).set(title="year vs duration")
plt.xticks(rotation=60)
```
